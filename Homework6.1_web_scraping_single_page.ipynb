```python
title=[]
artist=[]
release_time=[]
genre=[]
num_iter = len(soup.select('div.chart-content.col-xs-12.col-sm-8 > p '))
for i in range(num_iter):
    title.append(soup.select("cite.title")[i].get_text())
    artist.append(soup.select('em.artist')[i].get_text())
    release_time.append(soup.select('div.chart-content.col-xs-12.col-sm-8 > ul')[i].select('li')[-1].get_text().split(': ')[1])
for i in range(101):
    genre.append(soup.select('div.chart-content.col-xs-12.col-sm-8 > ul a')[i].get_text())
```
